```python
import pickle
from matplotlib import pyplot as plt
import glob
import os
import numpy as np

def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w

# first_round = 'data/fixed_full_track/frames_rewards282588.p'
# rewards_out1,frames_out1 = pickle.load( open(first_round, "rb" ) )

#p_files = sorted(glob.glob("data/fixed_full_track/frames_rewards288557.p"),key=os.path.getmtime)
p_files = sorted(glob.glob("data/ellipse_20210608/*.p"),key=os.path.getmtime)
ell_file = p_files[-1]
p_files = sorted(glob.glob("data/square_20210608/*.p"),key=os.path.getmtime)
sq_file = p_files[-1]
p_files = sorted(glob.glob("data/zigzag_20210608/*.p"),key=os.path.getmtime)
zig_file = p_files[-1]

print(ell_file)
print(sq_file)
print(zig_file)

mov_avg_len = 200
##elipse
rewards_ell,frames_out = pickle.load( open(ell_file, "rb" ) )
rewards_avg_ell = moving_average(rewards_ell,mov_avg_len)
##square
rewards_sq,frames_out = pickle.load( open(sq_file, "rb" ) )
rewards_avg_sq = moving_average(rewards_sq,mov_avg_len)
##zigzag
rewards_zig,frames_out = pickle.load( open(zig_file, "rb" ) )
rewards_avg_zig = moving_average(rewards_zig,mov_avg_len)

plot_raw = True
plot_sm = True

alpha = 0.3

plt.figure(dpi=100)
if plot_raw:
    plt.plot(rewards_sq,label='Sq',alpha=alpha)
if plot_sm:
    plt.plot(rewards_avg_sq,label='Sq Sm')
if plot_raw:
    plt.plot(rewards_ell,label='Ell',alpha=alpha)
if plot_sm:
    plt.plot(rewards_avg_ell,label='Ell Sm')
if plot_raw:
    plt.plot(rewards_zig,label='Zig',alpha=alpha)
if plot_sm:
    plt.plot(rewards_avg_zig,label='Zig Sm')
plt.legend()
plt.title('Episode Rewards ('+str(mov_avg_len)+' Episode Mov Avg)')
```
