```python
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
checkpoints = []
a = 1.5*np.sqrt(2)
rads = []
alphas = []
xs = []
ys = []
CHECKPOINTS = 12
for c in range(CHECKPOINTS):
    #self.np_random.seed(0)
    alpha = 2*math.pi*c/CHECKPOINTS# + self.np_random.uniform(0, 2*math.pi*1/CHECKPOINTS)
    #self.np_random.seed(0)
    #rad = self.np_random.uniform(TRACK_RAD/3, TRACK_RAD)
    if alpha <= 0.5*math.pi:
        rad = (a/np.sqrt(2))*(1/(np.sin(alpha)+np.cos(alpha)))
    elif alpha <= math.pi:
        rad = (a/np.sqrt(2))*(1/(np.sin(alpha-0.5*math.pi)+np.cos(alpha-0.5*math.pi)))
    elif alpha <= 1.5*math.pi:
        rad = (a/np.sqrt(2))*(1/(np.sin(alpha-math.pi)+np.cos(alpha-math.pi)))
    elif alpha <= 2*math.pi:
        rad = (a/np.sqrt(2))*(1/(np.sin(alpha-1.5*math.pi)+np.cos(alpha-1.5*math.pi)))
    checkpoints.append( (alpha, rad*math.cos(alpha), rad*math.sin(alpha)) )
    alphas.append(alpha)
    rads.append(rad)
    xs.append(np.cos(alpha)*rad)
    ys.append(np.sin(alpha)*rad)
```

```python
alpha = math.atan2(y, x)
while True: # Find destination from checkpoints
    failed = True
    while True:
        dest_alpha, dest_x, dest_y = checkpoints[dest_i % len(checkpoints)]
        if alpha <= dest_alpha:
            failed = False
            break
        dest_i += 1
        if dest_i % len(checkpoints) == 0: break
    if not failed: break
    alpha -= 2*math.pi
    continue
```
